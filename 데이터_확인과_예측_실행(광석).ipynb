```python
# 교차검증 연습

import pandas as pd
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score
```

```python
!git clone https://github.com/taehojo/data.git
```

```python
df = pd.read_csv('./data/sonar3.csv', header=None)
df
```

```python
x = df.iloc[:, :-1]
y = df.iloc[:, -1]
```

```python
k = 5
kfold = KFold(n_splits = k, shuffle=True)
acc_score = []
```

```python
# 모델 생성과 설정을 함수로 만듦
def model_fn():
  model = Sequential()
  model.add(Dense(32, input_dim=60, activation='relu'))
  model.add(Dense(16, activation='relu'))
  model.add(Dense(1, activation='sigmoid'))
  return model
```

```python
# k-fold 교차 검증을 이용해 k번 학습 실행
# for 문으로 k번 반복
# split()에 의해 k개의 학습셋, 테스트셋으로 분리
for train_idx, test_idx in kfold.split(x):
  xtrain, xtest = x.iloc[train_idx, :], x.iloc[test_idx, :]
  ytrain, ytest = y.iloc[train_idx], y.iloc[test_idx]

  model = model_fn()
  model.compile(loss = 'binary_crossentropy', optimizer='adam', metrics=['accuracy'])

  # verbose = 0으로 학습 과정의 출력을 생략
  history = model.fit(xtrain, ytrain, epochs=200, batch_size=10, verbose=0)

  accuracy = model.evaluate(xtest, ytest)[1]
  acc_score.append(accuracy)

# 텐서플로 함수에 for문을 포함하는 경우 WARNING 메세지가 나오는 경우가 있음, 그냥 무시 하면 됨
```

```python
# k번 실시된 정확도의 평균
avg_acc_score = sum(acc_score) / k

# 결과 출력
print('정확도:', acc_score)
print('평균 정확도:', avg_acc_score)
```
