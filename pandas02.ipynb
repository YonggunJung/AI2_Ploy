## 그룹별 집계 (Group by)

```python
import pandas as pd # pandas 모듈 호출
import numpy as np # numpy 모듈 호출

ipl_data = {'Team': ['Riders', 'Riders', 'Devils', 'Devils', 'Kings','kings', 'Kings', 'Kings', 'Riders', 'Royals', 'Royals', 'Riders'],
'Rank': [1, 2, 2, 3, 3,4 ,1 ,1,2 , 4,1,2],
'Year': [2014, 2015, 2014, 2015, 2014, 2015, 2016, 2017, 2016, 2014, 2015, 2017],
'Points':[876,789,863,673,741,812,756,788,694,701,804,690]}

df = pd.DataFrame(ipl_data)
df
```

```python
# df.groupby("묶을 열")["적용받는 열"].sum()
df.groupby("Team")["Points"].sum()
```

```python
# 멀티 인덱스 그룹별 집계.계층적 인덱스(hierarchical index) 형태
# 순서 있음
multi_groupby = df.groupby(["Team", "Year"])["Points"].sum()
multi_groupby
```

```python
multi_groupby.index
```

```python
# 순서 다르게. 결과는 똑같지만 형태가 다름
multi1_groupby = df.groupby(["Year", "Team"])["Points"].sum()
multi1_groupby
```

```python
multi1_groupby.index
```

```python
# 슬라이싱 가능
multi_groupby["Devils":"Riders"]
```

```python
# 없는건 NaN값으로 줌.
multi_groupby.unstack()
```

```python
multi_groupby
```

```python
multi_groupby.swaplevel()
```

```python
multi_groupby.swaplevel().sort_index()
```

```python
# level도 정해짐
multi_groupby.groupby(level = 0).sum()
```

```python
multi_groupby.groupby(level = 1).sum()
```

```python
grouped = df.groupby("Team")
grouped #그룹만 되어있지 어떻게 묶을지 안정해서 안보임
```

```python
# get_group 함수 : 해당 키 값을 기준으로 분할된 데이터프레임 객체를 확인
grouped.get_group("Riders")
```

```python
import warnings

# 모든 경고를 무시하는 필터를 설정합니다.
warnings.filterwarnings("ignore")
```

```python
# 집계(aggregation) : 요약된 통계 정보를 추출
# agg 함수 : min, 넘파이 mean 등 기존 함수 그대로 적용
# 모든 값의 최솟값으로 나옴
grouped.agg(min)
```

```python
grouped.agg(np.mean)
```

```python
# 변환(transformation) : 해당 정보를 변환
# 키 값별로 요약된 정보가 아닌 개별 데이터 변환 지원
# 적용 시점에서는 그룹화된 상태의 값으로 적용
# 모든 값의 최댓값으로 나옴
grouped.transform(max)
```

```python
score = lambda x: (x - x.mean()) / x.std()
grouped.transform(score)
```

```python
# 필터(filter) : 특정 조건으로 데이터를 검색
# 주로 filter 함수 사용
df.groupby('Team').filter(lambda x: len(x) >= 3)
```

```python
df.groupby('Team').filter(lambda x: x["Points"].max() > 800)
```

# 병합(merge)
## 두 개의 데이터를 특정 기준한 기준을 가지고 하나로 통합하는 작업

```python
import pandas as pd # pandas 모듈 호출
raw_data = {
'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
'test_score': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}

df_left = pd.DataFrame(raw_data, columns = ['subject_id', 'test_score'])
df_left
```

```python
raw_data = {
            'subject_id': ['4', '5', '6', '7', '8'],
            'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
            'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

df_right = pd.DataFrame(raw_data, columns = ['subject_id', 'first_name', 'last_name'])
df_right
```

```python
raw_data1 = {
            's_id': ['4', '5', '6', '7', '8'],
            'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
            'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

df_right1 = pd.DataFrame(raw_data1, columns = ['s_id', 'first_name', 'last_name'])
df_right1
```

```python
# 내부 조인(inner join) : 가장 기본적인 조인
# 집합으로 보면 양쪽의 교집합 데이터를 통합
pd.merge(left=df_left, right=df_right, how="inner", on='subject_id')
```

```python
pd.merge(left=df_left, right=df_right1, how="inner", on='subject_id')
```

```python
# 왼쪽 테이블과 오른쪽 테이블의 키 값이 다른 경우left_on과 right_on 매개변수를 사용하여 각 테이블 키 값을 입력
pd.merge(left=df_left, right=df_right1, left_on='subject_id', right_on='s_id')
```

```python
df_left.merge(df_right1, left_on='subject_id', right_on='s_id')
```

```python
# 왼쪽 조인 : 왼쪽 테이블을 기준으로 데이터를 병합
# 오른쪽 테이블에 왼쪽 테이블에 있는 키 값이 존재하지 않는다면 NaN으로 출력
pd.merge(df_left, df_right, on='subject_id', how='left')
```

```python
# 오른쪽 조인 : 오른쪽 테이블 기준으로 데이터를 병합
pd.merge(df_left, df_right, on='subject_id', how='right')
```

```python
# 완전 조인 : 두 테이블의 합집합을 의미
# 양쪽에 같은 키 값이 있는 데이터는 합치고 나머지는 NaN
pd.merge(df_left, df_right, on='subject_id', how='outer')
```

```python
df_left.index = df_left.subject_id
del df_left["subject_id"]
df_right.index = df_right.subject_id
del df_right["subject_id"]

pd.merge(df_left, df_right, on='subject_id', how='inner')
```
