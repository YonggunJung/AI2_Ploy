```python
import numpy as np
import pandas as pd

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
# 깃허브 데이터 가져오기
!git clone https://github.com/taehojo/data.git
```

```python
# 컬럼이 없음 그래서 데이터프레임 쓸 필요가 없음
df = pd.read_csv("./data/ThoraricSurgery3.csv")
df
```

```python
data = np.loadtxt("./data/ThoraricSurgery3.csv", delimiter=",")
data
```

```python
# 결과 데이터 분리
x = data[:, :-1]
y = data[:, -1]
```

```python
# 모델 생성
model = Sequential()

# 층을 만들고 구조 설정
# model.add()로 새로운 층 추가. 맨 마지막 층은 출력층, 나머지는 은닉층
# 이전 층에서 전달된 변수와 가중치, 바이어스가 하나로 모이는 곳
# Dense(노드 갯수, 독립변수 수,  활성함수)
# 데이터에서 값을 16개 입력받아 은닉층의 노드 30개로 보낸다는 의미
model.add(Dense(30, input_dim=16, activation='relu'))   # 은닉층
#출력 값을 하나로 정해야 하기 때문에 출력층 노드 수 1개
#활성 함수 시그모이드
model.add(Dense(1, activation='sigmoid'))   # 출력층

# 앞부분 내용을 컴퓨터가 알아 들을수 있게 컴파일
# 이항 분류은 loss는 binary_crossentropy 사용
# 다항 분류는 categorical_crossentropy
# 오차 조정해주는 옵티마이저는 보통 adam으로 생각하면 됨
# metrics()는 모델 수행 결과는 나타나게 설정
# accuracy는 정확도 기반, loss는 손실값, val_acc는 테스트셋 정확도, val_loss는 테스트셋 손실값
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
# batch_size는 샘플을 16개씩(행) 끊어서 집어 넣으라는 의미
history = model.fit(x, y, epochs=5, batch_size=16)
```
