```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

x1 = np.array([2, 4, 6, 8])
x2 = np.array([0, 4, 2, 3])
y = np.array([81, 93, 91, 97])
```

```python
# 데이터 분포 그래프
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter3D(x1, x2, y)
plt.show()
```

```python
# 값들 초기화
a1 = 0
a2 = 0
b = 0

# 학습률
lr = 0.01

# 반복횟구
epochs = 2001

# x1과 x2 수가 같아서 하나만 봄
n = len(x1)
```

```python

for i in range(epochs):
    y_pred = a1 * x1 + a2 * x2 + b
    error = y -y_pred

    a1d = (2/n) * sum(-x1 * error)
    a2d = (2/n) * sum(-x2 * error)
    bd = (2/n) * sum(-(error))

    a1 = a1 -lr * a1d
    a2 = a2 -lr * a2d
    b = b - lr * bd

    if i % 100 == 0:
        print("epoch=%.f, 기울기1 = %.04f, 기울기2 = %.04f, 절편 = %.04f" %(i, a1, a2, b))

print("실제 점수 : ", y)
print("예측 점수 : ", y_pred)
```
