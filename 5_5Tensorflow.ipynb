```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
x = np.array([2, 4, 6, 8])
y = np.array([81, 93, 91, 97])
```

```python
# 모델 생성
model = Sequential()
# 출력 값 입력 변수, 분석 방법에 맞게 모덜 설정
model.add(Dense(1, input_dim=1, activation='linear'))
# 오차 수정을 위한 optimizer, 오차 정도 판단을 위한 MSE
model.compile(optimizer='sgd', loss='mse')


model.fit(x, y, epochs=2000)

plt.scatter(x, y)
# 예측 결과 그래프
plt.plot(x, model.predict(x), 'r')
plt.show()
```

```python
# 임의 시간
hour = 7
# 임의 시간으로 점수 예측
pred = model.predict([hour])
print("%.f시간 공부하면 예상점수 : %.02f" %(hour, pred))
```

```python
x = np.array([[2,0],[4, 4], [6,2], [8,3]])
y = np.array([81, 93, 91, 97])
```

```python
# 모델 생성
model = Sequential()
# 출력 값 입력 변수, 분석 방법에 맞게 모덜 설정
model.add(Dense(1, input_dim=2, activation='linear'))
# 오차 수정을 위한 optimizer, 오차 정도 판단을 위한 MSE
model.compile(optimizer='sgd', loss='mse')


model.fit(x, y, epochs=2000)
```

```python
# 임의 시간
hour = 7
pric = 4

# 임의 시간으로 점수 예측
pred = model.predict([[hour,pric]])

print("%.f시간 공부하고, %.f시간 과외 예상점수 : %.02f" %(hour, pric, pred))
```

```python
x = np.array([[2,0, 1],[4, 4, 5], [6,2, 4], [8,3, 7]])
y = np.array([81, 93, 91, 97])
```

```python
# 모델 생성
model = Sequential()
# 출력 값 입력 변수, 분석 방법에 맞게 모덜 설정
model.add(Dense(1, input_dim=3, activation='linear'))
# 오차 수정을 위한 optimizer, 오차 정도 판단을 위한 MSE
model.compile(optimizer='sgd', loss='mse')


model.fit(x, y, epochs=2000)
```

```python
# 임의 값
hour = 7
pric = 4
th = 3
# 임의 시간으로 점수 예측
pred = model.predict([[hour,pric, th]])

print("%.f시간 공부하고, %.f시간 과외 예상점수 : %.02f" %(hour, pric, pred))
```
